```python
import numpy as np
from scipy.spatial import distance

customer_A = np.array([4, 5, 2, 3, 4])
customer_B = np.array([5, 3, 2, 4, 5])

customer_A_binary = np.array([1, 0, 1, 1, 0, 1])
customer_B_binary = np.array([1, 1, 1, 0, 0, 1])

euclidean_dist = np.linalg.norm(customer_A - customer_B)
print("1️ Euclidean Distance:", euclidean_dist)

manhattan_dist = np.sum(np.abs(customer_A - customer_B))
print("2️ Manhattan Distance:", manhattan_dist)

cosine_similarity = 1 - distance.cosine(customer_A, customer_B)
print("3️ Cosine Similarity:", cosine_similarity)

hamming_dist = distance.hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)
print("4️ Hamming Distance (binary):", hamming_dist)


A_likes = set(np.where(customer_A_binary == 1)[0])
B_likes = set(np.where(customer_B_binary == 1)[0])
jaccard_similarity = len(A_likes & B_likes) / len(A_likes | B_likes)
print("5 Jaccard Similarity (binary likes):", jaccard_similarity)


user1 = np.array([5, 3, 4, 4, 2])
user2 = np.array([4, 2, 5, 4, 3])

chebyshev_dist = distance.chebyshev(user1, user2)
print("\n6️ Chebyshev Distance:", chebyshev_dist)

minkowski_dist = distance.minkowski(user1, user2, p=3)
print("7️ Minkowski Distance (p=3):", minkowski_dist)
```
