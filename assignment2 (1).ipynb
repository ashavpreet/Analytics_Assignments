```python
#questio8
employees = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

emp_df = pd.DataFrame(employees)
# a) Shape
print("\nQ8.a) Shape of DataFrame:", emp_df.shape)
# b) Summary (info)
print("\nQ8.b) DataFrame Summary:")
print(emp_df.info())
# c) Descriptive Statistics
print("\nQ8.c) Descriptive Statistics:")
print(emp_df.describe())

# d) Display first 5 and last 3 rows
print("\nQ8.d) First 5 rows:")
print(emp_df.head())
print("\nLast 3 rows:")
print(emp_df.tail(3))

# e) Statistics
avg_salary = emp_df['Salary'].mean()
total_bonus = emp_df['Bonus'].sum()
youngest = emp_df['Age'].min()
highest_rating = emp_df['Rating'].max()

print("\nQ8.e) Calculated Statistics:")
print("Average Salary:", avg_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest)
print("Highest Rating:", highest_rating)

# f) Sort by Salary descending
sorted_df = emp_df.sort_values(by='Salary', ascending=False)
print("\nQ8.f) Sorted by Salary (Descending):")
print(sorted_df)

# g) Categorize by performance rating
def categorize(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

emp_df['Performance_Category'] = emp_df['Rating'].apply(categorize)
print("\nQ8.g) Performance Categories:")
print(emp_df[['Name', 'Rating', 'Performance_Category']])

# h) Identify missing values
print("\nQ8.h) Missing Values:")
print(emp_df.isnull().sum())

# i) Rename Employee_ID to ID
emp_df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\nQ8.i) Column Renamed:")
print(emp_df.head())

# j) Filter employees
more_than_5_years = emp_df[emp_df['Years_of_Experience'] > 5]
it_department = emp_df[emp_df['Department'] == 'IT']

print("\nQ8.j.i) Employees with >5 years experience:")
print(more_than_5_years)

print("\nQ8.j.ii) Employees in IT department:")
print(it_department)

# k) Add new column 'Tax' = 10% of Salary
emp_df['Tax'] = emp_df['Salary'] * 0.10
print("\nQ8.k) Added Tax column:")
print(emp_df[['Name', 'Salary', 'Tax']])

# l) Save modified DataFrame to CSV
emp_df.to_csv('Modified_Employees.csv', index=False)
print("\nQ8.l) Modified DataFrame saved to 'Modified_Employees.csv'")
```
