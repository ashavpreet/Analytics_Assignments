```python
import pandas as pd
customers = pd.read_csv("AWCustomers.csv")
sales = pd.read_csv("AWSales.csv")
```

```python
merged = pd.merge(customers, sales, on='CustomerID', how='inner')
merged.columns
```

```python
selected_cols = ["Education","BirthDate","NumberCarsOwned","Occupation","MaritalStatus","HomeOwnerFlag","TotalChildren","YearlyIncome","BikeBuyer"]

selected_df = merged[selected_cols]
selected_df.head()
```

```python
avg_income = selected_df.groupby('BikeBuyer')['YearlyIncome'].mean()
avg_income
```

```python
percentage_bikebuyers = selected_df.groupby('Occupation')['BikeBuyer'].mean() * 100
print(percentage_bikebuyers)
```

```python
from datetime import datetime
import pandas as pd

# Ensure proper conversion
selected_df['BirthDate'] = pd.to_datetime(selected_df['BirthDate'], errors='coerce')

# Drop or handle invalid dates if any
selected_df = selected_df.dropna(subset=['BirthDate'])

# Now compute Age
selected_df['Age'] = datetime.now().year - selected_df['BirthDate'].dt.year

selected_df.head()
```
